```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Ensure plots look nice
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
def simulate_generation(current_freq, N, fitness_A=1.0, fitness_a=1.0):
    """
    N: Population size (number of individuals, each with 2 alleles)
    fitness_A, fitness_a: Relative fitness of alleles
    """
    total_alleles = 2 * N
    # Calculate probability of picking allele A given selection
    prob_A = (current_freq * fitness_A) / (current_freq * fitness_A + (1 - current_freq) * fitness_a)

    # Random sampling of alleles for the next generation
    new_count_A = np.random.binomial(total_alleles, prob_A)
    return new_count_A / total_alleles
```

```python
def run_simulation(N=100, generations=100, initial_freq=0.5, fitness_A=1.0, fitness_a=1.0, replicates=10):
    plt.figure(figsize=(10, 6))

    for i in range(replicates):
        freqs = [initial_freq]
        for gen in range(generations):
            new_f = simulate_generation(freqs[-1], N, fitness_A, fitness_a)
            freqs.append(new_f)
        plt.plot(freqs, alpha=0.5)

    plt.xlabel('Generation')
    plt.ylabel('Allele Frequency (A)')
    plt.title(f'Evolutionary Simulation (N={N}, Replicates={replicates})')
    plt.ylim(0, 1)
    plt.show()
```

```python
interact(run_simulation,
         N=widgets.IntSlider(min=10, max=1000, step=10, value=100),
         generations=widgets.IntSlider(min=10, max=500, step=10, value=100),
         initial_freq=widgets.FloatSlider(min=0.01, max=0.99, step=0.01, value=0.5),
         fitness_A=widgets.FloatSlider(min=0.8, max=1.2, step=0.05, value=1.0),
         replicates=widgets.IntSlider(min=1, max=50, step=1, value=10));
```

```python
def simulate_generation(current_freq, N, fitness_A, fitness_a, mut_a_to_A=0.001, mut_A_to_a=0.001):
    # 1. Apply Mutation
    # Frequency after mutation, but before selection/drift
    freq_after_mut = current_freq * (1 - mut_A_to_a) + (1 - current_freq) * mut_a_to_A

    # 2. Selection
    # Calculate probability of picking allele A based on fitness
    total_fitness = (freq_after_mut * fitness_A) + ((1 - freq_after_mut) * fitness_a)
    prob_A = (freq_after_mut * fitness_A) / total_fitness

    # 3. Drift (Sampling)
    new_count_A = np.random.binomial(2 * N, prob_A)
    return new_count_A / (2 * N)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets  # This is the line that defines 'interact'
```

```python
def run_simulation(N=100, generations=100, initial_freq=0.5, fitness_A=1.0, fitness_a=1.0,
                   mut_a_to_A=0.001, mut_A_to_a=0.001, replicates=10):
    plt.figure(figsize=(10, 6))

    for i in range(replicates):
        freqs = [initial_freq]
        for gen in range(generations):
            current_f = freqs[-1]

            # 1. Apply Mutation
            freq_after_mut = current_f * (1 - mut_A_to_a) + (1 - current_f) * mut_a_to_A

            # 2. Selection
            total_fitness = (freq_after_mut * fitness_A) + ((1 - freq_after_mut) * fitness_a)
            prob_A = (freq_after_mut * fitness_A) / total_fitness

            # 3. Drift (Sampling)
            new_f = np.random.binomial(2 * N, prob_A) / (2 * N)
            freqs.append(new_f)

        plt.plot(freqs, alpha=0.5)

    plt.xlabel('Generation')
    plt.ylabel('Allele Frequency (A)')
    plt.title(f'Evolutionary Simulation (N={N}, Replicates={replicates})')
    plt.ylim(0, 1)
    plt.show()
```

```python
interact(run_simulation,
         N=widgets.IntSlider(min=10, max=1000, step=10, value=100),
         generations=widgets.IntSlider(min=10, max=500, step=10, value=100),
         initial_freq=widgets.FloatSlider(min=0.01, max=0.99, step=0.01, value=0.5),
         fitness_A=widgets.FloatSlider(min=0.8, max=1.2, step=0.05, value=1.0),
         fitness_a=widgets.FloatSlider(min=0.8, max=1.2, step=0.05, value=1.0),
         mut_a_to_A=widgets.FloatSlider(min=0, max=0.05, step=0.001, value=0.001),
         mut_A_to_a=widgets.FloatSlider(min=0, max=0.05, step=0.001, value=0.001),
         replicates=widgets.IntSlider(min=1, max=50, step=1, value=10));
```
